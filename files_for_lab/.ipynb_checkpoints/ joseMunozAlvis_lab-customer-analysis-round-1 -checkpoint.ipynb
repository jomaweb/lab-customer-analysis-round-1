### Delete the column education  and the number of open complaints from the dataframe.

### Correct the values in the column customer lifetime value. 
They are given as a percent, 
so multiply them by 100 and change dtype to numerical type.

```python
# https://datatofish.com/string-to-integer-dataframe/
# https://www.geeksforgeeks.org/how-to-convert-string-to-integer-in-pandas-dataframe/
# https://stackoverflow.com/questions/13682044/remove-unwanted-parts-from-strings-in-a-column
# https://datatofish.com/replace-character-pandas-dataframe/

#df['column name'] = df['column name'].str.replace('old character','new character')
# df['first_set'] = df['first_set'].str.replace('_','|')
```

```python
# repalce the unnecessary characters 

new_data["customer lifetime value"] = new_data["customer lifetime value"].str.replace("%", "")
```

```python
#Replace NaN Values with Zeros
# https://datatofish.com/replace-nan-values-with-zeros/
#df['DataFrame Column'] = df['DataFrame Column'].fillna(0)
new_data["customer lifetime value"] = new_data["customer lifetime value"].fillna(0)
```
